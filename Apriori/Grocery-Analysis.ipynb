```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules

dataset = pd.read_csv('./filename.csv',header=None)
data = []
for i in range(len(dataset)):
    data.append([j for j in dataset.values[i] if j == j])
    
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)

freq_itemsets = apriori(df, min_support=0.33, use_colnames=True)
my_rules = association_rules(freq_itemsets,min_threshold=0.6)

my_rules['length'] = my_rules['antecedents'].apply(lambda x: len(x)) + 
                     my_rules['consequents'].apply(lambda x: len(x))
    
my_rules = my_rules[my_rules['length']==3] # replace 3 by maximum length in your numerical calculation
my_rules[["antecedents","consequents","confidence"]]
```
